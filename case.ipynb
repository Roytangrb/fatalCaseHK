```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib notebook
```

```python
df = pd.read_csv("wiki.csv")

df['Date'] = pd.to_datetime(df['Date'])

df_50_17 = df[(df['Date'].dt.year >= 1950) & (df['Date'].dt.year <= 2017)]

count = df_50_17['Date'].groupby(df['Date'].dt.year).size()
#print(count) 1967:26cases/ 2002:43cases
'''
manually add case without exact day
    [1962, 1974, 1979, 1980, 1981, 1985, 1988, 1989, 1991, 1998, 2001, 2002, 2003, 2008]
    1977: 3
    1982: 3
    1983: 2
    1984: 3
    1992: 2
    1993: 2
    1995: 3
    1996: 4
    1997: 5
    1999: 3  
'''
add_one = [1962, 1974, 1979, 1980, 1981, 1985, 1988, 1989, 1991, 1998, 2001, 2002, 2003, 2008]
for year in add_one:
    count.loc[year] = count.loc[year] + 1

count.loc[1977] = count.loc[1977] + 3
count.loc[1982] = count.loc[1982] + 3
count.loc[1983] = count.loc[1983] + 2
count.loc[1984] = count.loc[1984] + 3
count.loc[1992] = count.loc[1992] + 2
count.loc[1993] = count.loc[1993] + 2
count.loc[1995] = count.loc[1995] + 3
count.loc[1996] = count.loc[1996] + 4
count.loc[1997] = count.loc[1997] + 5
count.loc[1999] = count.loc[1999] + 3 

#plot
sns.set(style="darkgrid")
plt.plot(count, '-o', markerSize=3);
plt.scatter([1967, 2002], [26, 44], s=30, c='red')
plt.xticks(list(np.arange(1950, 2018, 10)) + [2017]);
plt.yticks(list(np.arange(0, 50, 5)))
plt.title("Reported Cases Involving Killing, 1950 - 2017, Hong Kong");
#annotate
ax1 = plt.gca()
ax1.annotate('1967', xy=(1967, 26), xytext=(1967, 27), ha='center');
ax1.annotate('2002', xy=(2002, 43), xytext=(2002.5, 44.5), ha='center');
```
